Task5:- Visualizing High-Dimensional Data

Generate a parallel coordinates plot for visualizing high-dimensional data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("100_Sales.csv").drop(columns=['Unnamed: 9','Unnamed: 10'])
data
```

```python
data_for_plot = data[['Item_Type', 'Unit_Cost', 'Total_Revenue', 'Total_Profit']]
```

```python
from pandas.plotting import parallel_coordinates
plt.figure(figsize=(14, 7))
parallel_coordinates(data_for_plot, 'Item_Type', colormap=plt.get_cmap("Set3"))
```
