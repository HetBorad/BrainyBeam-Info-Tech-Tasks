Task 2:- Data Manipulation

Use broadcasting to normalize a 2D numpy array row-wse..

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("100_Sales.csv").drop(columns=['Unnamed: 9','Unnamed: 10'])
data
```

```python
# features for normalization
features = data[['Unit_Cost', 'Total_Revenue']].values
```

```python
row_sums = features.sum(axis=1, keepdims=True) 
```

```python
normalized_features = features / row_sums
```

```python
print("Original two Features were Unit_Cost and Total_Revenue:\n", features)
print("\nRow Sums:\n", row_sums)
print("\nNormalized Features (Row-wise):\n", normalized_features)
```
